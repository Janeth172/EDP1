```python
# IMPORTAMOS LIBRERIAS
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
```

SE DEFINE LA FUNCIÓN A TROZOS

```python
# ESCRIBIR EL PRIMER TROZO
def f1(x):                    # SE DEFINE UNA FUNCIÓN, COMO ES UNA FUNCIÓN SOLO SE MANDA A LLAMRA
    y = zeros(len(x))            # TODAS LAS ENTRADAS DE Y SERAN CEROS, len(x) ES CUANTAS ENTRADAS TIENE x
    for i in range(len(x)):
        y[i] = 0
    return y
def f2(x):                    # SE DEFINE UNA FUNCIÓN, COMO ES UNA FUNCIÓN SOLO SE MANDA A LLAMRA
    y = zeros(len(x))            # TODAS LAS ENTRADAS DE Y SERAN CEROS, len(x) ES CUANTAS ENTRADAS TIENE x
    for i in range(len(x)):
        y[i] = 2-x[i]
    return y
```

```python
x1 = np.linspace(-2,0,1000)  # linspace (valor-inicial, valor-final, número de valores) 1000 SON MUESTRAS O PUNTOS UNIDOS
x2 = np.linspace(0,2,1000)
y1 = f1(x1)     # f MANDA A LLAMAR LA FUNCION
y2 = f2(x2)
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label=r'$f(x) = x + 2$ para $-2 \leq x \leq 0$')
plt.plot(x2, y2, label=r'$f(x) = 0$ para $0 x \leq x \leq 2$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');    # NOMBRA LOS EJES
plt.title("Condición inicial (u(x,0)=f(x))")
xlim([-3.0,3.01]),ylim([-1.05,3.01]),grid(),legend()     # xlim y ylim es el cuadrito
show()
```

```python
# SERIE DE FOURIER
def u_0(x,N=40): # N es el número de armónicos, ENTRE MAS ARMONICOS MAS NOS APROXIMAMOS
    f = zeros(len(x))      # LA FUNCIÓN LA LLENA DE CEROS, DEPENDIENDO DE LA LONGITUD DE X
    for n in range(1,N):   # n inicializado en 0, empieza en 1 hasta 20
        f += ((2-2*(-1)**n)/(np.pi**2 * n**2) * np.cos((n*np.pi*x)/2) + (2/(np.pi*n)) * np.sin((n*np.pi*x)/2)) # += ES UNA SUMATORIA, este f es la serie de Fourier
    return (1/2)+f  # ES EL 1/2 a0
```

```python
x = np.concatenate((x1, x2))
f_n = u_0(x)
plot(x1, y1, label='$f(x) = 0$ para $-2 \leq x \leq 0$') # GRAFICA f1
plot(x2, y2, label='$f(x) = 2-x$ para $0 x \leq x \leq 2$')   # GRAFICA f2
plot(x, f_n, label='$f_N(x)$ con $N=40$')                 # GRAFICA LA FUNCION
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2.0,2.1]),ylim([-0.1,3.2]),grid(),legend()
show()
```

```python
# SE CULPLE EL TEOREMA DE CONVERGENCIA PUNTUAL, PUES COMO LA FUNCION POR LA DERECHA CONVERGE A DOS Y POR LA IZQUIERDA CONVERGE
# A 0 ENTONCES SU PROMEDIO ES 1 POR EL TEOREMA YA DICHO DE 1/2 (F(-X) + F(X+))
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)   # ES LA GRAFICA DE CADA FUNCION USANDO DIFERENTES ARMONICOS
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2.2,3.01]),ylim([-0.5,3.2]),grid(),legend()
show()
```
