```python
from sympy import fourier_series, pi, plot, Piecewise, E # IMPORTAMOS E
from sympy.abc import x # SE DEFINIRA X COMO VARIABLE
```

**EN COSENOS, ES DECIR PAR**

```python
f=Piecewise(       # SE DEFINE UNA FUNCION ATROZOS
    (E**(2*x), (x > 0) & (x <= 1)), # USE E en ves de  e
    (E**(-2*x), (x >= -1) & (x <= 0)) # USE E enves de e
)
```

```python
plot(f, (x,-1,1),line_color ='red')
```

```python
s = fourier_series(f, (x, -1, 1))      # SE CALCULA LA SERIE DE FOURIER f(x) EN EL INTERVALO -1,1
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```

**EN SENOS, ES DECIR IMPAR**

```python
Y=Piecewise(       # SE DEFINE UNA FUNCION ATROZOS
    (E**(2*x), (x > 0) & (x <= 1)), # USE E en ves de e
    ((-1)*E**(-2*x), (x >= -1) & (x <= 0)) # USE E en ves de e
)
```

```python
plot(Y, (x,-1,1),line_color ='red')
```

```python
ss = fourier_series(Y, (x, -1, 1))      # SE CALCULA LA SERIE DE FOURIER f(x) EN EL INTERVALO -1,1
ss1 = ss.truncate(n = 3)
ss2 = ss.truncate(n = 5)
ss3 = ss.truncate(n = 7)
p = plot(Y, ss1, ss2, ss3, (x, -pi, pi), show=False, legend=True)
```

```python
ss
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```

**SERIE DE FOURIER EN COSENOS Y SENOS DE LA FUNCION A TROZOS**

```python
L=Piecewise(       # SE DEFINE UNA FUNCION ATROZOS
    (1, (x >= 0) & (x <= (pi/2))), # USE E en ves de e
    (2, (x > (pi/2)) & (x <= pi)) # USE E en ves de e
)
```

```python
plot(L, (x,-2,pi),line_color ='red')
```

```python
sss = fourier_series(L, (x, 0, pi))      # SE CALCULA LA SERIE DE FOURIER EN SENOS f(x) EN EL INTERVALO -1,1
sss1 = sss.truncate(n = 3)
sss2 = sss.truncate(n = 5)
sss3 = sss.truncate(n = 7)
pt = plot(L, sss1, sss2, sss3, (x, 0, pi), show=False, legend=True)
```

```python
sss
```

```python
pt[0].label = 'x'
pt[1].label = 'n=3'
pt[2].label = 'n=5'
pt[3].label = 'n=7'
```

```python
pt.show()
```
